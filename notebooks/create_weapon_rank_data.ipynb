# Cleaning up the Weapon Rank Column

Throwing together this quick notebook to build a new data file for weapon ranks. I really didn't like how I originally set it up and after mucking about in PowerBI for a while I realized that's because this data should be stored in a long format on its own. I'm thankfully realizing more about how I want my semantic model to look and this data definitely proves the value of using relational data going forward.

I'm going to do more comprehensive data cleaning later as my data in PowerBI looks quite different to this now, but this will do for now.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/the-sacred-stones/full_df.csv")
df
```

```python
df.iloc[0]
```

```python
import ast

df_small = df[["Name", "Weapon Rank", "creature_campaign"]]
df_small.head(15)
```

```python
df_small
```

```python
from ast import literal_eval

def handle_ranks(ranks):
    """
    Functions goal is to take the weapon rank column like "["A Sword", "B Axe"]"
    and convert it into a format I can actually work with. 
    
    Function also standardizes the missing weapon ranks.
    """
    if (isinstance(ranks, str)) and (ranks != "–"):
        ranks = literal_eval(ranks)
    elif (isinstance(ranks, float) and np.isnan(ranks)) or ranks == "–":
        ranks = []  
    return ranks
```

```python
df_small["Weapon Rank"] = df_small["Weapon Rank"].apply(handle_ranks)
```

```python
df_small.iloc[0]["Weapon Rank"]
```

```python
df_small = df_small.explode("Weapon Rank").dropna()
```

```python
df_small
```

```python
df_small[['Rank', 'Weapon']] = df_small['Weapon Rank'].str.split(' ', expand=True)
df_small.drop(columns = "Weapon Rank", inplace=True)
```

```python
df_small.to_csv("../data/the-sacred-stones/weapon_ranks.csv")
```
